```python
import pandas as pd
import numpy as np
from zipfile import ZipFile
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier

path = '/home/ragav/my_python_files/folder_data/diabetes.zip'
myzip = ZipFile(path)
zz = myzip.namelist()
print(zz)
myfile = myzip.open(zz[0])
# read the data in
df = pd.read_csv(myfile)
#print(df.info())
print(df.head())

# summarize the number of unique values in each column
print(df.nunique())
# dataset
dataset = df.values
# load pima indians dataset
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
y = dataset[:,8]

# summarize the shape of the training dataset
print(X.shape, y.shape)
# identify outliers in the training dataset
isof = IsolationForest()
yhat = isof.fit_predict(X)
# select all rows that are not outliers
mask = yhat != -1
X, y = X[mask, :], y[mask]
# summarize the shape of the updated training dataset
print(X.shape, y.shape)

# perform a robust scaler transform of the dataset
trans = StandardScaler()
X = trans.fit_transform(X)


# RFE
# define the method
rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=4)
# fit the model
rfe.fit(X, y)
# transform the data
X = rfe.transform(X)

print(X.shape)
print(y.shape)

np.savetxt('X.txt', X, delimiter=',')   # X is an array
np.savetxt('y.txt', y, delimiter=',')   # X is an array
## save the isolation forest model
#dump(isof, open('isof.pkl', 'wb'))
# save the scaler
#dump(trans, open('trans.pkl', 'wb'))
print('~~~~~')
```
